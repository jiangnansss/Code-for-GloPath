```python
import sys
import os
notebook_path = os.getcwd()
sys.path.append(os.path.join(os.path.dirname(notebook_path), '0-Models_and_weights', '1-Segmentor'))
sys.path.append(os.path.join(os.path.dirname(notebook_path), '0-Models_and_weights', '0-GloPath'))
from glopath_segmentor import loadGloSeg, valTrans, quantification
import glob
import torch
from tqdm import tqdm
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Init Segmentor

```python
bowSegmentor = loadGloSeg(structure='bow').cuda().eval()
tuftSegmentor = loadGloSeg(structure='tuft').cuda().eval()
```

Segmentation & quantification

```python
gloNames = glob.glob('/data_sdd/hqm/xj/Code-for-GloPath/2-Data/0-examples/*')

with torch.no_grad():
    for gloName in tqdm(gloNames):
        print(gloName)
        gloImg = cv2.imread(gloName)[...,::-1]
        h, w= gloImg.shape[:2]
        gloTensor = valTrans(image=gloImg)['image'].unsqueeze(0).cuda()
        bowMask = bowSegmentor(gloTensor)[0].argmax(0).detach().cpu().numpy().astype(np.uint8)*255
        tuftMask = tuftSegmentor(gloTensor)[0].argmax(0).detach().cpu().numpy().astype(np.uint8)*255
        bowMask = cv2.resize(bowMask, (w,h), fx=0, fy=0)
        tuftMask = cv2.resize(tuftMask, (w,h), fx=0, fy=0)
        quantRes = quantification(bowMask, tuftMask)
        print(quantRes)
        zeros = np.zeros_like(bowMask)
        bowMask = np.stack((bowMask, zeros, zeros), 2)
        tuftMask = np.stack((tuftMask, zeros, zeros), 2)
        bowMask = (gloImg*0.5+bowMask*0.5).astype(np.uint8)
        tuftMask = (gloImg*0.5+tuftMask*0.5).astype(np.uint8)
        
        plt.figure(figsize=(9,3))
        plt.subplot(131)
        plt.imshow(gloImg)
        plt.subplot(132)
        plt.imshow(bowMask)
        plt.subplot(133)
        plt.imshow(tuftMask)
        plt.show()
```
